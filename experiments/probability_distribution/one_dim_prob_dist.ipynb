```python
import scipy.stats as stats

from my_stats.probability_distribution.probability_distribution import plot_distribution
```

# Discrete Probability Distribution 

## Binominal Distribution

ベルヌーイ試行の成功回数を表す分布。

```python
rv = stats.binom(n=10, p=0.3)
plot_distribution(rv, title="Binomial Distribution")
```

## Poisson distribution

稀に起こる事象が一定時間内に起こる回数を表す確率分布。
np一定でnを無限大にした時の二項分布の極限となる分布。

```python
rv = stats.poisson(mu=3)
plot_distribution(rv, title="Poisson Distribution")
```

```python
rv = stats.poisson(mu=10)
plot_distribution(rv, title="Poisson Distribution")
```

# Continual Probability Distribution 

## Normal Distribution

最も重要な確率分布。
中心極限定理より任意の確率分布の和は正規分布に収束する。

```python
rv = stats.norm(loc=0, scale=1)
plot_distribution(rv, title="Standard Normal Distribution")
```

## Parato Distribution

```python
rv = stats.pareto(b=1)
plot_distribution(rv, title="Pareto Distribution")
```

## Laplace distribution(double exponential distribution)

```python
rv = stats.laplace()
plot_distribution(rv, title="Laplace Distribution")
```
