# Discrete Probability Distribution 

## Binominal Distribution

ベルヌーイ試行の成功回数を表す分布。

```python
rv = stats.binom(n=10, p=0.3)
plot_distribution(rv, title="Binomial Distribution")
```

## Poisson distribution

稀に起こる事象が一定時間内に起こる回数を表す確率分布。
np一定でnを無限大にした時の二項分布の極限となる分布。

```python
rv = stats.poisson(mu=3)
plot_distribution(rv, title="Poisson Distribution")
```

```python
rv = stats.poisson(mu=10)
plot_distribution(rv, title="Poisson Distribution")
```

# Continual Probability Distribution 

## Normal Distribution

最も重要な確率分布。
中心極限定理より任意の確率分布の和は正規分布に収束する。

```python
rv = stats.norm(loc=0, scale=1)
plot_distribution(rv, title="Standard Normal Distribution")
```

# Exponential Distribution(指数分布)

# Exponential distribution PDF
### Probability Density Function (PDF)

The PDF of the exponential distribution is:

$$f(x; \lambda) = \lambda e^{{-\lambda x}} \quad \text{{for }} x \geq 0$$

where:
- $\lambda$ is the rate parameter (λ > 0)
- $x$ is the random variable



正の値のみに確率をもつ分布。
生存時間の分布として用いられることがある。
無記憶性を持つ。

```python
# Scipyではlambdaはscaleの逆数で指定する
lambda_param = 3
rv = stats.expon(scale=1 / lambda_param)
plot_distribution(rv, title="Exponential Distribution")
```

```python
lambda_param = 0.5
rv = stats.expon(scale=1 / lambda_param)
plot_distribution(rv, title="Exponential Distribution")
```
