# 確率積分変換

```python
import scipy.stats as stats
from my_stats.probability_integral_transform import visualize_pit
```

# 正規分布

```python
rv = stats.norm(loc=0, scale=1)
print(type(rv))
visualize_pit(rv, n_samples=10000)
```

# ベータ分布

```python
# beta distribution
rv = stats.beta(a=2, b=5)
visualize_pit(rv, n_samples=10000)
```

# ガンマ分布

```python
# gamma distribution
rv = stats.gamma(a=2, scale=2)
visualize_pit(rv, n_samples=10000)
```
